```python
import numpy as np
import matplotlib.pyplot as plt
```

# The cost function 
$ J = (1.1 - x- sin(x))^2 $

```python
plt.style.use(['ggplot'])
def  cal_cost(X):
    cost = (np.square(1.1-(X+np.sin(X))))
    return cost 
```

# Derivative of the cost function
$\Delta J = 2*(1.1-x-sin(x))*(-1-cos(x))$

```python
def gradient_descent(learning_rate=0.01,iterations=100):
   
    cost_history = np.zeros(iterations)
    x_hat = 2
    cost_history[0] = x_hat

    for it in range(iterations):
        
        x_hat = x_hat - learning_rate*(2*(1.1 - x_hat - np.sin(x_hat))*(-1 - np.cos(x_hat)))
        
        cost_history[it]  = cal_cost(x_hat)
        
        it = it + 1
        
    return cost_history
```

# Test with different learning rates: 0.005, 0.01, 0.05, 0.1

```python
lr =0.005
n_iter = 500

cost_history = gradient_descent(lr,n_iter)

print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
fig,ax = plt.subplots(figsize=(10,8))
_=ax.plot(range(200),cost_history[:200],'b.')
```

```python
lr =0.01
n_iter = 500

cost_history = gradient_descent(lr,n_iter)

print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
fig,ax = plt.subplots(figsize=(10,8))
_=ax.plot(range(200),cost_history[:200],'b.')
```
